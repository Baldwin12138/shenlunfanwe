## 利用xpath解析


```python
from lxml import etree
```

### 解析目录页

```python
menuselector = etree.HTML(content_str, parser=etree.HTMLParser(encoding='gbk'))
```

```python
# 文章链接
st = menuselector.xpath(r'//div[@class="zg_main"]')
stt = st[0].xpath(r'div[@class="zg_mainbox_let fl"]/ul[@class="list"]') 
sttt = stt[0].xpath(r'li/a')
print(sttt[0].attrib)
words = menuselector.xpath(r'//div[@class="zg_main"]/div[@class="zg_mainbox_let fl"]/ul[@class="list"]/li/a')
print(words[0].attrib)
print(words[-1].attrib)
```

```python
# 下一页目录链接
nextpage = menuselector.xpath(r'//div[@class="zg_main"]/div[@class="zg_mainbox_let fl"]/div[@class="zg_page list_pagebox"]\
/p/a[text()="下一页"]')[0]
print(nextpage.attrib)
nextpage2 = menuselector.xpath(r'//div[@class="zg_main"]/div[@class="zg_mainbox_let fl"]/div[@class="zg_page list_pagebox"]\
/p/a[7]')[0]
print(nextpage2.text)
print(nextpage2.attrib["href"])
nextpage3 = menuselector.xpath(r'//div[@class="zg_main"]/div[@class="zg_mainbox_let fl"]/div[@class="zg_page list_pagebox"]\
/p/a[7]/@href')
print(nextpage3)
```

### 解析文章页

```python
wordselector = etree.HTML(content_stra, parser=etree.HTMLParser(encoding='gbk'))
```

```python
wt = wordselector.xpath(r'//div[@class="zg_main"]/div[@class="zg_left"]')[0]
print(wt)
# 标题
title = wt.xpath(r'h1[@class="zg_show_tit"]')[0]
print(title.text)
# 发布日期
date = wt.xpath(r'div[@class="zg_show_form"]')[0]
print(date.text.replace('|','').strip())
# 正文
word = wt.xpath(r'div[@class="zg_show_word"]/p/text()') # list每个元素为一段落
print('\t'+'\n\t'.join(word)) 
```
